```python
import shutil

def copy_directory(src_dir, dest_dir):
    try:
        shutil.copytree(src_dir, dest_dir,dirs_exist_ok=True)
        print("Directory copied successfully!")
    except FileExistsError:
        print(f"Error: Destination directory '{dest_dir}' already exists.")
    except PermissionError:
        print(f"Error: Permission denied to access '{dest_dir}'.")
    except Exception as e:  # Catch any other unexpected errors
        print(f"Error during copying: {e}")
source_directory = "./drive/MyDrive/DATN/DiffuSum"
destination_directory = "./DiffuSum"
copy_directory(source_directory, destination_directory)
```

```python
import zipfile

def extract_zip(zip_file, extract_dir):
    try:
        with zipfile.ZipFile(zip_file, 'r') as zip_ref:
            zip_ref.extractall(extract_dir)
        print("Extraction successful!")
    except FileNotFoundError:
        print(f"Error: ZIP file '{zip_file}' not found.")
    except zipfile.BadZipFile:
        print(f"Error: '{zip_file}' is not a valid ZIP file.")

zip_file_path = "./drive/MyDrive/ACL2020_data.zip"
extract_to = "./DiffuSum/datasets/cnndm"
extract_zip(zip_file_path, extract_to)

zip_file_path = "./drive/MyDrive/ACL2020_other_datasets.zip"
extract_to = "./DiffuSum/datasets/others"
extract_zip(zip_file_path, extract_to)
```
